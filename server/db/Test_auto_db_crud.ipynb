```python
f_data = list()
for i in df_file.index:
    for file in df_file[i]:
        if not file['mode']=='hidden_by_limit':
            fid = file['id']
            f_link = file['url_private_download']
            f_data.append([fid,f_link])
f_df = pd.DataFrame(f_data,columns=['fid','f_link'])
f_df
```

```python
def download(url, path='../data/rsc/', file_name = None):
    if not os.path.isdir(path):
        os.mkdir(path)
    
    if not file_name:
        file_name = path + '/' + url.split('/')[-1].split('?')[0]


    with open(file_name, "wb") as file:   
            response = req.get(url)               
            file.write(response.content)
    return file_name
```

```python
file_list = list()
path='../data/rsc/'
for fid,f_link in zip(f_df['fid'],f_df['f_link']):
    file_name = download(f_link,path + fid)
    file_list.append(file_name)
```
